```python
import pandas as pd   
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
import matplotlib.pyplot as plt
#Data reading, merging, random splitting.
unames = ['user_id','gender','age','occupation','zip'] 
users = pd.read_table('users.dat',sep='::',header= None,names= unames,engine='python') 
rnames = ['user_id','movie_id','rating','timestamp'] 
ratings = pd.read_table('ratings.dat',sep='::',header=None,names=rnames,engine='python')
mnames=['movie_id','title','genres'] 
movies = pd.read_table('movies.dat',sep='::',header=None,names=mnames,engine='python') 
data_1=pd.merge(pd.merge(ratings,users),movies)  
train = data_1.sample(n=None, frac=0.8, replace=False, weights=None, random_state=42, axis=None)
test = data_1.drop(train.index)
```

```python
#The experimental design is based on age as a fixed effect, gender and occupation as random effects.
genres_set = ['Action','Adventure','Animation',"Children's",'Comedy','Crime','Documentary','Drama','Fantasy','Film-Noir','Horror','Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western']
y = []
for i in range(len(genres_set)):
    train_i = train[train['genres'].str.contains(genres_set[i])]
    data = train_i
    mod_lme = smf.mixedlm("rating~0+C(age)",data=data,groups=data['movie_id'],re_formula="1",vc_formula={"gender": "0+C(gender)", "occupation": "0+C(occupation)"})
    result = mod_lme.fit()
    #Study the influence of different ages on the ratings of various types of movies and draw radar maps.
    a = list(result.params)
    for k in range (0,3):
        a.pop()
    labels = np.array(["Under 18","18-24","25-34","35-44","45-49","50-55","56+"])
    dataLenth = 7
    data2 = np.array(a)
    angles = np.linspace(0, 2*np.pi, dataLenth, endpoint=False)
    data2 = np.concatenate((data2, [data2[0]])) 
    angles = np.concatenate((angles, [angles[0]])) 
    fig = plt.figure()
    ax = fig.add_subplot(111, polar=True)
    ax.plot(angles, data2, 'ro-', linewidth=1)
    ax.set_thetagrids(angles * 180/np.pi, labels, fontproperties="SimHei")
    ax.set_title(genres_set[i], va='bottom', fontproperties="SimHei")
    ax.grid(True)
    plt.show()
    #Evaluate the accuracy of the LMM model based on MAE.
    test_i = test[test['genres'].str.contains(genres_set[i])]
    test_i = test_i.reset_index(drop = True)
    pred_i = result.predict(test_i)
    error_i = []
    for j in range(len(test_i)):
        error_i.append(test_i.ix[[j]].values[0][2] - pred_i[j])
    absError_i = []
    for val in error_i:
        absError_i.append(abs(val))
    print(genres_set[i],"MAE = ", sum(absError_i) / len(absError_i))
    y.append(sum(absError_i) / len(absError_i))
y
```
